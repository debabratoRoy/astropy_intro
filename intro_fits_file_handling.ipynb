```python
from astropy.io import fits
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
fits_file = fits.open('hlsp_hugs_hst_wfc3-uvis_ngc1851_f275w_v1_stack-1288s.fits')
print(fits_file)
```

```python
image_data = fits_file[0].data
print(image_data)
```

```python
plt.figure()
plt.imshow(image_data, origin = 'lower', norm=colors.LogNorm())
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.imshow(image_data, origin = 'lower', norm=colors.LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```

```python
section1 = image_data[2250:2650, 5350:5950]
```

```python
plt.imshow(section1,origin='lower',norm=colors.LogNorm())
plt.colorbar()
plt.show()
```
