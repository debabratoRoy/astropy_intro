```python
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
fits_ngc1261 = fits.open('hlsp_hugs_hst_wfc3-uvis_ngc1261_f275w_v1_stack-0859s.fits')
print(fits_ngc1261)
```

```python
img_data = fits_ngc1261[0].data
print(img_data)
```

```python
plt.imshow(img_data,origin='lower',norm=colors.LogNorm())
plt.colorbar()
plt.show()
```

```python
plt.imshow(img_data,origin='lower',cmap='Greys',norm=colors.LogNorm())
plt.colorbar()
plt.show()
```

```python
sec1 = img_data[2250:2650, 5350:5950]
```

```python
plt.imshow(sec1,origin='lower',norm=colors.LogNorm())
plt.colorbar()
plt.show()
```

```python
plt.imshow(sec1,origin='lower',norm=colors.LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```
